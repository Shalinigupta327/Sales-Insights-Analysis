```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

## EDA

```python
# plotting a bar chart for Gender and it's count

a = sns.countplot(x = 'Gender', data = df)
for bars in a.containers:
    a.bar_label(bars)
```

```python
# plotting a bar chart for gender vs total amount
purchase = df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending =False)
sns.barplot(x = 'Gender', y= 'Amount', data = purchase)
```

## Age

```python
a = sns.countplot(x = 'Age Group', data = df, hue= 'Gender')
for bars in a.containers:
    a.bar_label(bars)
```

```python
# Total Amount vs Age Group
sales_age = df.groupby(['Age Group'], as_index= False)['Amount'].sum().sort_values(by= 'Amount', ascending= False)
sns.barplot(x= 'Age Group', y= 'Amount', data= sales_age)
```

## State

```python
df.columns
```

```python
sales_state= df.groupby(['State'], as_index= False)['Orders'].sum().sort_values(by= 'Orders', ascending= False).head(9)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x= 'State', y= 'Orders', data= sales_state)
```

```python
sales_state = df.groupby(['State'], as_index= False)['Amount'].sum().sort_values(by= 'Amount', ascending= False).head(9)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x= 'State', y= 'Amount', data= sales_state)
```

## Marital Status

```python
a = sns.countplot(x= 'Marital_Status', data= df)
sns.set(rc={'figure.figsize':(8,5)})
for bars in a.containers:
    a.bar_label(bars)
```

```python
sales_mar= df.groupby(['Marital_Status', 'Gender'], as_index= False)['Amount'].sum().sort_values(by= 'Amount', ascending= False)
sns.barplot(x= 'Marital_Status', y= 'Amount', data= sales_mar, hue= 'Gender')
```
